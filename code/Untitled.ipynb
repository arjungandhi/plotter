```python
import numpy as np 
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('jeremy.jpg')
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
```

```python
#simplifying image
num_color  = 3
div = int((255+1)/num_color)
print(div)
simp_img = img // div * div + div // 2
simp_img2 = cv2.cvtColor(simp_img, cv2.COLOR_BGR2RGB)
plt.imshow(simp_img2)
```

```python
channels = cv2.split(img)
num_colors = 8
new_channels = [np.zeros_like(c).astype('float64') for c in channels]
r = 255/num_colors
for i in range(num_colors):
        lower = r*(i-1)
        upper = r*i
        val = (lower+upper)/2
        for i,c in enumerate(channels):
             new_channels[i] += np.where(abs(c-val) <=  r , val, 0)

new_channels = [c.astype('uint8') for c in new_channels]
new_img = cv2.merge(new_channels)
new_img2 = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img2)
```
